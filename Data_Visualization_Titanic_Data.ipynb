```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=snr.load_dataset('titanic')
```

```python
data.head()
```

```python
#histogram
data['fare'].diff().plot(kind='hist', bins=200)
```

```python
#histogram
data['age'].diff().plot(kind='hist', bins=200)
```

```python
#scatter plot
data.plot.scatter(x='fare', y='age', c='pclass')
```

```python
snr.relplot(x='fare', y='age', data=data, hue='who')
```

```python
snr.relplot(x='fare', y='age', data=data, hue='who', style='alive')
```

```python
snr.relplot(x='fare', y='age', data=data, hue='who', style='alive', col='alone')
```

```python
snr.relplot(x='fare', y='age', data=data, hue='who', style='alive', col='alone', size='sibsp')
```

```python
#boxplot
snr.catplot(x='who', y='age', hue='sibsp', kind='box', data=data)
```
